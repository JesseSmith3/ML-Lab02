```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
bostonHouseFrame=pd.read_csv('s3://rmit-ml-2020s1-lab-data/BostonHousingPrice/housing.data.csv', delimiter='\s+')
```

```python
bostonHouseFrame.shape
```

```python
print(bostonHouseFrame)
```

```python
pd.DataFrame.max(bostonHouseFrame)
```

```python
pd.DataFrame.min(bostonHouseFrame)
```

```python
pd.DataFrame.mean(bostonHouseFrame)
```

```python
pd.DataFrame.median(bostonHouseFrame)
```

```python
bostonHouseFrame.describe()
```

```python
plt.figure()
bostonHouseFrame['RM'].hist(figsize=(4,4))
plt.show()
plt.figure()
bostonHouseFrame['LSTAT'].hist(figsize=(4,4))
plt.show()
plt.figure()
bostonHouseFrame['MEDV'].hist(figsize=(4,4))
plt.show()
```

```python
bostonHouseFrame.plot(kind='box', subplots=True, layout=(4,4), sharex=False, sharey=False, figsize=(20,20))
plt.show
```

```python
import seaborn as sns

correlations = bostonHouseFrame.corr()

f, ax = plt.subplots(figsize=(11,9))

sns.heatmap(correlations, cmap="PuBu", center=0, vmin=-1, vmax=1, square=True)

b, t = plt.ylim()
b += 0.5
t -= 0.5
plt.ylim(b, t)
plt.show()
```

```python
plt.scatter(bostonHouseFrame['RM'], bostonHouseFrame['MEDV'])
plt.xlabel('Average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupier homes in $1000s')
plt.show()

plt.scatter(bostonHouseFrame['LSTAT'], bostonHouseFrame['MEDV'])
plt.xlabel('% lower status of the population')
plt.ylabel('Median value of owner-occupied homes in $1000s')
plt.show()
```

```python
plt.figure()
bostonHouseFrame["RM","LSTAT","MEDV"].hist(figsize=(20,20))
plt.show()
```
