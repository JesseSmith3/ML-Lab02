```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
```

```python
bostonHouseFrame=pd.read_csv('s3://rmit-ml-2020s1-lab-data/BostonHousingPrice/housing.data.csv', delimiter='\s+')
bostonHouseFrame
```

```python
X = bostonHouseFrame[['LSTAT']]
Y = bostonHouseFrame[['MEDV']]
```

```python
from sklearn.model_selection import train_test_split

trainX, testX, trainY, testY = train_test_split(np.array(X), np.array(Y), test_size=0.2)
print(trainX.shape)
print(testX.shape)
print(trainY.shape)
print(testY.shape)
```

```python
from sklearn import linear_model
linReg = linear_model.LinearRegression()
```

```python
linReg.fit(trainX,trainY)
```

```python
print(linReg.intercept_)
print(linReg.coef_)
```

```python
predYRm = linReg.predict(testX)
```

```python
from sklearn.metrics import mean_squared_error
print('Mean squared error ', mean_squared_error(testY, predYRm))
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
plt.figure()

plt.subplot(1,2,1)
plt.scatter(trainX, trainY)
plt.plot(trainX,linReg.predict(trainX),color='red',linewidth=4.0)
plt.xlabel("LSTAT")
plt.ylabel("MEDV")
plt.title('Train Data')

plt.subplot(1,2,2)
plt.scatter(testX, testY)
plt.plot(testX,linReg.predict(testX),color='red',linewidth=4.0)
plt.xlabel("LSTAT")
plt.ylabel("MEDV")
plt.title('Test Data')
plt.show()
```

```python
plt.scatter(testX, linReg.predict(testX) - testY, c='g', s=40)
plt.hlines(y=0, xmin=3.5, xmax=40)
```

```python
X = bostonHouseFrame.drop('MEDV', axis=1)
X = X.drop('CHAS', axis=1)
X = X.drop('DIS', axis=1)
Y = bostonHouseFrame[['MEDV']]
```

```python
X
```

```python
trainX, testX, trainY, testY = train_test_split(np.array(X), np.array(Y), test_size=0.2)
print(trainX.shape)
print(testX.shape)
print(trainY.shape)
print(testY.shape)
```

```python
linReg = linear_model.LinearRegression()
linReg.fit(trainX,trainY)
print(linReg.intercept_)
print(linReg.coef_)
```

```python
predYRm = linReg.predict(testX)
```

```python
print('Mean squared error ', mean_squared_error(testY, predYRm))
```

```python
from sklearn.preprocessing import PolynomialFeatures
polyFeat = PolynomialFeatures(degree=2, include_bias=True)
polyTrainX = polyFeat.fit_transform(trainX)
polyTestX = polyFeat.fit_transform(testX)
linReg.fit(polyTrainX,trainY)
```

```python
predYRm = linReg.predict(polyTestX)
```

```python
print('Mean squared error ', mean_squared_error(testY, predYRm))
```
