```python
#set up feature/attribute data AND the column which is TARGET_5Yrs
X = nba.drop('TARGET_5Yrs', axis=1)
X = X.drop('Name', axis=1)
Y = nba['TARGET_5Yrs']

#split into nba into test/train, then split train into train/validation
from sklearn.model_selection import train_test_split
#first split
nbaTrainX, nbaTestX, nbaTrainLabels, nbaTestLabels = train_test_split(X, Y, test_size=0.2)
#second split
nbaTrainX, nbaValX, nbaTrainLabels, nbaValLabels = train_test_split(nbaTrainX, nbaTrainLabels, test_size=0.2)


print(nbaTrainX.shape)
print(nbaTrainLabels.shape, "\n")

print(nbaTestX.shape)
print(nbaTestLabels.shape, "\n")

print(nbaValX.shape)
print(nbaValLabels.shape)
```

```python
#NOTE: Evaluation methods will be accuracy and F1-score

# Logistic regression model 
# C specifies the amount of weighting placed on minimising the loss function (error).
# The implementation of logistic regression in Scikit learn includes regularisation, something to prevent overfitting.
# We donâ€™t use regularisation, so C is set to a large value, which effectively negates the regularisation.

# The max_iter is the maximum number of iterations to fit parameters to the model, as the
# Logistics regressor uses a method similar to gradient descent. Although typically not a parameter that we tune
# too much, to demonstrate how parameter tuning works we will try various settings of max_iter to find the best
# setting (for the training/validation data).

from sklearn.linear_model import LogisticRegression
logRegNba = LogisticRegression(C=10000, max_iter=100, solver='liblinear')

logRegNba.fit(nbaTrainX, nbaTrainLabels)
pred = logRegNba.predict(nbaTestX)
print("Test Predicion: ", pred)

from sklearn.metrics import f1_score, accuracy_score
print(f1_score(nbaTestLabels, pred))
print(accuracy_score (nbaTestLabels, pred))
```

```python
# PARAMETER TUNING

lMaxIter = [19,50,100,250,1000]
lF1Score = []
lAccuracyScore = []

for maxIter in lMaxIter:
    currLogRegNba = LogisticRegression(C=1, max_iter=maxIter, solver='liblinear')
    currLogRegNba.fit(nbaTrainX, nbaTrainLabels)
    currPred = currLogRegNba.predict(nbaValX)
    lF1Score.append(f1_score(nbaValLabels, currPred))
    lAccuracyScore.append(accuracy_score(nbaValLabels, currPred))
    
print(lF1Score)
print(lAccuracyScore)

plt.rcParams['figure.figsize'] = [16, 8]
plt.figure()
plt.subplot(1,2,1)
plt.plot(lMaxIter, lF1Score)
plt.ylabel('F1 Score')
plt.xlabel('max iter')

plt.subplot(1,2,2)
plt.plot(lMaxIter, lAccuracyScore)
plt.ylabel('Accuracy')
plt.xlabel('max iter')

plt.show()
```

```python
currLogRegNba = LogisticRegression(C=100000, max_iter=76, solver='liblinear')
currLogRegNba.fit(nbaTrainX, nbaTrainLabels)
currPred = currLogRegNba.predict(nbaTestX)
print(f1_score(nbaTestLabels, currPred))
print(accuracy_score(nbaTestLabels, currPred))
```
